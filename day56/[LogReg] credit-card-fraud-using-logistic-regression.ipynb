## Importing modules and datasets

```python
# For algorithm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# For preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Check the distribution of legit and fraudulent transactions

```python
# Check the distribution of legit and fraudulent transactions
data['Class'].value_counts()
```

From the output the data is far from balanced, to make it balance we should do this

* 0 --> Normal Transaction
* 1 --> Fraudulent Transaction

```python
# Separating data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

## Under-Sampling

Build a sample dataset containing similar distributions of normal transactions and fraudulent transactions

Fraudulent transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

## Concatenating two dataframes

```python
df_new = pd.concat([legit_sample, fraud], axis=0)
```

```python
df_new.head()
```

```python
df_new.tail()
```

```python
df_new['Class'].value_counts()
```

```python
df_new.groupby('Class').mean()
```

It's far more better right than before taking sample in legit value

## Splitting the data into features and targets

```python
X = df_new.drop(columns='Class', axis=1) # features
y = df_new['Class'] # target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

### Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

## Model Evaluation

### Accuracy Score

```python
# Accuracy on train data
X_train_prediction = model.predict(X_train)
data_train_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy on training data:", data_train_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
data_test_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print("Accuracy on test data:", data_test_accuracy)
```

If accuracy on training data and on test data is far different, perhaps the model is underfitting or overfitting

But in the model that I build, the different is not too far
