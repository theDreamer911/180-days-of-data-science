## Import module

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Canada.xlsx', 'Canada by Citizenship', 
                   skiprows=range(20), skipfooter=2)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

## Data Preprocessing/Cleaning/Wrangling

```python
df.drop(['AREA', 'Type', 'Coverage', 'REG', 'DEV'], axis=1, inplace=True)
```

### Renaming Features

```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df.head()
```

```python
## Adding a total feature
df['Total'] = df.sum(axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.Country
```

```python
df[['Country', 1980, 1981, 1982, 1983, 1984, 1985]]
```

```python
df.set_index('Country', inplace=True)
df.head()
```

### Convert column name to string

```python
df.columns = list(map(str, df.columns))
```

```python
df.columns
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
print("Data dimensions:", df.shape)
print(df.columns)
```

```python
haiti = df.loc['Haiti', years] # Passing year 1980 - 2013 exclude 'total' columns
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

## Area plot

```python
df.sort_values(['Total'], ascending=False, inplace=True)
```

```python
df_top5 = df.head()
df_top5 = df_top5[years].head()
df_top5
```

```python
df_top5.plot(kind='area', alpha=0.35, stacked=False, figsize=(30,20))
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

## Histograms

```python
df['2013'].head()
```

```python
count, bin_edges = np.histogram(df['2013'])
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind='hist')
plt.title("Histogram of Immigrations from 195 countries in 2013")
plt.ylabel("Number of Countries")
plt.xlabel("Number of Immigrants")
plt.show()
```

## Bar charts

```python
df_iceland = df.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6))
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.title("Icelandic Immigrants to Canada from 1980 to 2013")
plt.show()
```

## Pie Charts

```python
df_continents = df.groupby('Continent', axis=0).sum()
df_continents['Total'].plot(kind='pie')
plt.title("Immigration to Canada by Continent 1980-2013")
plt.show()
```

## Box Plots

```python
df_japan = df.loc[['Japan'], years].transpose()
df_japan.plot(kind='box')
plt.title("Box plot of Japannese Immigration from 1980-2013")
plt.ylabel("Number of Immigrants")
plt.show()
```

## Scatter Plot

```python
df_total = pd.DataFrame(df[years].sum(axis=0))
df_total.index = map(int, df_total.index)
df_total.reset_index(inplace=True)
df_total.columns=['years', 'Total']
df_total
```

```python
df_total.plot(kind='scatter', x='years', y='Total')
plt.title("Total Immigration to Canada from 1980 - 2013")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
```

## Regression Plots

```python
ax = sns.regplot(x='years', y='Total', data=df_total, color='green', marker="*")
```

## Maps with markers

```python
canada_map = folium.Map(location=[56.130, -106.35], zoom_start=4)
```

```python
# Ontario contains about 40% of Canadian Populations
# Adding circular mask to the center of Ontario
ontario = folium.map.FeatureGroup()
ontario.add_child(folium.CircleMarker([51.25, -85.32], radius=5, color='red', fill_color='Red'))

# Add feature group to the map
canada_map.add_child(ontario)

# Label the marker
folium.Marker([51.25, -85.32], popup='ontario').add_to(canada_map)
```
