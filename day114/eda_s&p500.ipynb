# S&P 500 Analysis

Source : [Wikipedia](https://en.wikipedia.org/wiki/List_of_S%26P_500_companies)

Tutorial : [Data Professor](https://www.youtube.com/watch?v=JwSS70SZdyM)

```python
import pandas as pd

def load_data():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    html = pd.read_html(url, header=0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

## Retrieve the Stock Data

```python
data = yf.download(
    tickers = list(df.Symbol),
    period = "ytd",
    interval = "1d",
    groupby = 'ticker',
    auto_adjust = True,
    prepost = True, 
    threads = True, 
    proxy = None    
)
```

```python
data
```

```python
data.columns = data.columns.swaplevel(0,1)
data.sort_index(axis=1, level=0, inplace=True)
```

```python
data
```

## Data Subset of Query Symbol

```python
data['ABT']
```

```python
df2 = pd.DataFrame(data['ABT'].Close)
df2['Date'] = df2.index 
df2
```

## Plot Closing Price of Query Symbol

```python
import matplotlib.pyplot as plt
plt.fill_between(df2.Date, df2.Close, color="skyblue", alpha=0.3)
plt.plot(df2.Date, df2.Close, color="skyblue", alpha=0.8)
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Closing Price")
```

```python
import matplotlib.pyplot as plt

def plot_price(symbol):
    df = pd.DataFrame(data[symbol].Close)
    df['Date'] = df.index
    plt.fill_between(df.Date, df.Close, color="skyblue", alpha=0.3)
    plt.plot(df.Date, df.Close, color="skyblue", alpha=0.8)
    plt.xticks(rotation=90)
    plt.title(symbol, fontweight="bold")
    plt.xlabel("Date", fontweight="bold")
    plt.ylabel("Closing Price", fontweight="bold")
    return plt.show()
```

```python
plot_price('AAPL')
```

```python
for i in list(df.Symbol)[:10]:
    plot_price(i)
```
