```python
import matplotlib.pyplot as plt
import numpy as np
import pygal
```

## Matplotlib Radar Chart

```python
# Data
categories = ['HP', 'MP', 'Attack', 'Defense', 'Speed']
categories = np.concatenate((categories, [categories[0]]))

# Player
# Warrior
player1 = [180, 78, 127, 110, 80]
player1 = np.concatenate((player1, [player1[0]]))

# Monk
player2 = [160, 150, 99, 90, 60]
player2 = np.concatenate((player2, [player2[0]]))

# Knight
player3 = [200, 50, 117, 120, 130]
player3 = np.concatenate((player3, [player3[0]]))
```

```python
player1
```

```python
# Calculate evenly-spaced angle coordinates
# Using radians for polar plot with 2*np.pi
label_placement = np.linspace(start=0, stop=2*np.pi, num=len(player1))
```

```python
label_placement
```

```python
print(2*np.pi, 'radians =', np.degrees(2*np.pi), 'degrees')
```

```python
# Radians
print('radians', label_placement)
# Degrees
print('degrees', np.degrees(label_placement))
```

```python
## Visualize the radar chart
# Declare empty plot chart
plt.figure(figsize=(8,8))
plt.subplot(polar=True)

# Ploting each player
plt.plot(label_placement, player1)
plt.plot(label_placement, player2)
plt.plot(label_placement, player3)

# Coloring
# plt.fill(label_placement, player1, 'red', alpha=0.1)
# plt.fill(label_placement, player2, 'orange', alpha=0.1)
# plt.fill(label_placement, player3, 'blue', alpha=0.1)

# Thetarids to place labels at specified angles using degrees
lines, labels = plt.thetagrids(np.degrees(label_placement), labels=categories)
plt.title("Compare Players Stats", y=1.1, fontdict={"fontsize":24})
plt.legend(labels=['Warrior', 'Wizard', 'Knight'], loc=(0.95, 0.8))
```
