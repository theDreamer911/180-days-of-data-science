```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd
lego_data = pd.read_csv('datasets/lego_sets.csv')
```

```python
lego_data.shape
lego_data.info()
lego_data.head()
```

```python
lego_data_clean = lego_data.dropna(subset=['set_num'])
lego_data_clean.info()
```

```python
parent_data = pd.read_csv('datasets/parent_themes.csv')
parent_data.info()
parent_data.head(10)
```

```python
licensed_data = parent_data[parent_data['is_licensed'] == True]['name']
print(licensed_data.head(10))
type(licensed_data)
```

```python
licensed = lego_data_clean['parent_theme'].isin(licensed_data)
```

```python
licensed_set = lego_data_clean[licensed]
print(licensed_set['parent_theme'].unique())
licensed_set.head()
```

```python
lego_combined = lego_data_clean.merge(parent_data,
                                      how='left',
                                      left_on='parent_theme',
                                      right_on='name',
                                      suffixes=('', '_y')
                                     )
```

```python
lego_data_clean.head()
```

```python
lego_combined.head()
```

```python
lego_combined_licensed = lego_combined[lego_combined['is_licensed'] == True]
licensed_sets_all = lego_combined_licensed.drop(axis=1, columns=['id', 'name_y', 'is_licensed'])
licensed_sets_all.equals(licensed_set)
```
